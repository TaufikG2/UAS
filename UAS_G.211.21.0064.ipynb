```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv('Stocks.csv')
df
```

```python
plt.figure(figsize=(10, 4))

plt.subplot(1, 3, 1)
sns.histplot(df['Purchase Price ($)'], kde=True, bins=13, edgecolor='w', color="#ff4d4d")
plt.xlabel('Purchase Price')
plt.ylabel('Frequency')
plt.title('Histogram of Purchase Price')

plt.subplot(1, 3, 2)
sns.histplot(df['Current Price ($)'], kde=True, bins=11, edgecolor='w', color="#b366ff")
plt.xlabel('Current Price')
plt.ylabel('Frequency')
plt.title('Histogram of Current Price')

plt.subplot(1, 3, 3)
sns.histplot(df['Gain/Loss ($)'], kde=True, bins=13, edgecolor='w', color="#ff0066")
plt.xlabel('Gain/Loss')
plt.ylabel('Frequency')
plt.title('Histogram of Gain/Loss')

plt.tight_layout()
plt.show()
```

```python
data = df.sort_values(by="Dividend/Shares", ascending=False)

top_5 = data.head(5)
#plot data
plt.figure(figsize=(5, 5))
sns.barplot(x="Stock Name", y="Dividend/Shares", data=top_5, color="#ffcc99")
plt.title("Top 5 Stocks with Highest Dividend / Shares")
plt.show()
```

```python
data = df.sort_values(by="Market Value ($)", ascending=False)

top_5 = data.head(5)
#plot data
plt.figure(figsize=(5, 5))
sns.barplot(x="Stock Name", y="Market Value ($)", data=top_5, color="#cc66ff")
plt.title("Top 5 Stocks With Highest Market Value ($)")
plt.show()
```

```python
data = df.sort_values(by="Annual Yield (%)", ascending=False)

top_7 = data.head(7)
#plot data
plt.figure(figsize=(13, 10))
sns.barplot(x="Stock Name", y="Annual Yield (%)", data=top_7, color="#66ccff")
plt.title("Top 7 Stocks with Highest Annual Yield ($)")
plt.show()
```
